```python
from pandas_datareader import data
from matplotlib import pyplot as plt
import pandas as pd
import datetime
import numpy as np

print(np.__version__)
```

```python
#define the instruments to download

companies_dict = {
    'Amazon': 'AMZN',
    'Apple': 'AAPL',
    'Walgreen': 'WBA',
    'Northrop Grumman': 'NOC',
    'Boeing': 'BA',
    'Lockheed Martin': 'LMT',
    'McDonalds': 'MCD',
    'Intel': 'INTC',
    'Navistar': 'NAV',  
    'IBM': 'IBM',
    'Texas Instruments': 'TXN',
    'MasterCard': 'MA',
    'Microsoft': 'MSFT',
    'General Electrics': 'GE',
    'Symantec': 'SYMC',
    'American Express': 'AXP',
    'Pepsi': 'PEP',
    'Coca Cola': 'KO',
    'Johnson & Johnson': 'JNJ',
    'Toyota': 'TM',
    'Honda': 'HMC',
    'Mitsubishi': 'MSBHY',
    'Sony': 'SNE',
    'Exxon': 'XOM',
    'Chevron': 'CVX',
    'Valero Energy': 'VLO',
    'Ford': 'F',
    'Bank of America': 'BAC'
}
companies = sorted(companies_dict.items(), key=lambda x: x[1])
print(companies)
```

```python
#define which online source to use
data_source = 'yahoo'

#define the start and end dates
start_data = '2015-01-01'
end_data = '2017-12-31'

#using th pandas reader to lload the desired data
panel_data = data.DataReader(list(companies_dict.values()), data_source, start_data, end_data)

#print the axes labels
print(panel_data.axes)
```

```python
#find the stock open and close data
stock_close = panel_data['Close']
stock_open = panel_data['Open']

print(stock_close.iloc[0])
```

```python
# calculating daily stock movement
stock_close = np.array(stock_close).T
stock_open = np.array(stock_open).T

print(stock_close.shape)

row, col = stock_close.shape
print(row, col)
```

```python
movements = np.zeros([row, col])

for i in range(0, row):
    movements[i, :] = np.subtract(stock_close[i, :], stock_open[i, :])
    
for i in range(0, len(companies)):
    print("Company: {}, Change: {}".format(companies[i][0], sum(movements[i][:])))
```

```python
# pre processing the data
from sklearn.preprocessing import Normalizer

#create the normaliser
normalizer = Normalizer()
new = normalizer.fit_transform(movements)

print(new.max())
print(new.min())
print(new.mean())
```

```python
#visualising the data 
plt.clf
plt.figure(figsize=(18,16))
ax1 = plt.subplot(221)
plt.plot(new[0][:])
plt.title(companies[0])


plt.subplot(222, sharey=ax1)
plt.plot(new[1][:])
plt.title(companies[1])

plt.show()
```

```python
# make the pipeline
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer

# Define the normaliser
normalizer = Normalizer()

 #create a KMeans model
kmeans = KMeans(n_clusters=10, max_iter=1000)

#make a pipeline
pipeline = make_pipeline(normalizer, kmeans)
```

```python
#fit pipeline to daily stock movements
pipeline.fit(movements)

#how packed are the cluster to each other
print(kmeans.inertia_)
```

```python
# import pandas
import pandas as pd

# predict the cluster label
labels = pipeline.predict(movements)


# create a dataframe to visualise
df = pd.DataFrame({'labels': labels, 'companies': companies})

#display df sorted by cluster_label
print(df.sort_values('labels'))
```
